```python
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('display.max_rows', None)
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
df = pd.read_csv('../data/cleaned/clean_2012_2024.csv')
```

```python
df.corr(numeric_only=True)
```

```python
# Age
plt.figure(figsize=(12,4))
feature_to_plot = 'surfer_age'
sns.histplot(df[feature_to_plot], kde=True)
plt.show()

sns.catplot(x='surfer_age', y='surfer_won', data=df, kind='bar', aspect=2)
plt.xticks(rotation=90)
plt.title('Heat Place Distribution')
plt.xlabel('Age')
plt.ylabel('Average Heat Place')
plt.show()
```

```python
# Stance
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='surfer_stance', y='surfer_win_rate')
plt.title('Win Rate by Stance')
plt.xlabel('Stance')
plt.ylabel('Win Rate')
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='surfer_frontside', y='surfer_won')
plt.title('Heat Place Distribution by Frontside')
plt.xticks([0, 1], ['False', 'True'])
plt.xlabel('Frontside')
plt.ylabel('Average Heat Place')
plt.show()
```

```python
# Country
plt.figure(figsize=(12,4))

plt.figure(figsize=(12,4))
sns.countplot(data=df, x='surfer_country', order=df['surfer_country'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Number of Surfers by Country')
plt.ylabel('Number of Surfers')
plt.xlabel('Country')
plt.tight_layout()
plt.show()


# Calculate win rate per country
country_win_rate = df.groupby('surfer_country')['surfer_won'].mean().reset_index()
plt.figure(figsize=(12,4))
sns.barplot(x='surfer_country', y='surfer_won', data=country_win_rate.sort_values('surfer_won', ascending=False))
plt.xticks(rotation=90)
plt.title('Win Rate by Surfer Country')
plt.xlabel('Country')
plt.ylabel('Win Rate')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12,4))
sns.catplot(x='surfer_country', y='surfer_won', data=df, kind='bar', aspect=4, height=4)
plt.xticks(rotation=90)
plt.title('Heat Place Distribution By Country')
plt.xlabel('Age')
plt.ylabel('Average Heat Place')
plt.show()
```

```python
# Correlations
plt.figure(figsize=(12, 4))
sns.scatterplot(data=df, x='surfer_avg_ht', y='surfer_win_rate', hue='surfer_gender')
plt.title('Correlation of Average Heat Score and Win Rate by Gender')
plt.xlabel('Average Heat Score')
plt.ylabel('Win Rate')
plt.legend(title='Gender')
plt.show()
```

```python
selected_columns = df[['surfer_points', 'surfer_age', 'surfer_win_rate', 'surfer_avg_ht']]
sns.pairplot(selected_columns)
plt.show()
```

```python
num_cols = ['surfer_won', 'surfer_avg_ht', 'surfer_points', 'surfer_age', 'surfer_win_rate', 'surfer_heat_wins', 'surfer_heat_losses']
correlation_matrix = df[num_cols].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='surfer_won', y='surfer_age', data=df)
plt.title('Age Distribution by Win Status')
plt.xlabel('Won')
plt.ylabel('Age')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='surfer_won', y='surfer_avg_ht', data=df)
plt.title('Average Heat Score Distribution by Win Status')
plt.xlabel('Won')
plt.ylabel('Average Heat Score')
plt.show()
```

```python
plt.figure(figsize=(16, 6))
sns.kdeplot(data=df[df['surfer_won'] == 1]['surfer_points'], label='Winners', shade=True)
sns.kdeplot(data=df[df['surfer_won'] == 0]['surfer_points'], label='Non-Winners', shade=True)
plt.title('Surfer Points Distribution for Winners vs. Non-Winners')
plt.xlabel('CT. Points for Current Year')
plt.ylabel('Density')
plt.legend()
plt.show()

plt.figure(figsize=(16, 6))
sns.kdeplot(data=df[df['surfer_won'] == 1]['surfer_prev_5_event_points'], label='Winners', shade=True)
sns.kdeplot(data=df[df['surfer_won'] == 0]['surfer_prev_5_event_points'], label='Non-Winners', shade=True)
plt.title('Surfer Points Distribution for Winners vs. Non-Winners')
plt.xlabel('Total CT. Points for Prior 5 Events')
plt.ylabel('Density')
plt.legend()
plt.show()
```
